Predicting diabetes
In the previous project we saw how we could use a decision tree to predict data sets related to classification and regression. However, did you know that we can improve the prediction of a tree by using a random forest?

As we have studied, a random forest is a grouping of trees generated with random portions of the data and with random criteria. This view would allow us to improve the effectiveness of the model when an individual tree is not sufficient.

In this project improve accuracy.

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python

data_train = pd.read_csv(rf"C:\Users\wipip\OneDrive\Documentos\GitHub\Random_Forest_Tree_Project-main\data\processed\data_train.csv")
data_test = pd.read_csv(rf"C:\Users\wipip\OneDrive\Documentos\GitHub\Random_Forest_Tree_Project-main\data\processed\data_test.csv")
print(data_train.shape, data_test.shape)
print(data_train.columns, data_test.columns)
```

```python
X_train = data_train.drop('diagnositco', axis=1)
y_train = data_train['diagnositco']
X_test = data_test.drop('diagnositco', axis=1)
y_test = data_test['diagnositco']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
mod_0 = RandomForestClassifier(random_state=42)
mod_0.fit(X_train, y_train)
```

```python
y_pred_0 = mod_0.predict(X_test)
y_pred_0
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, classification_report


print(f'El accuracy de nuestro modelo en forma simple es: {accuracy_score(y_test, y_pred_0)} y la metrica roc_auc {roc_auc_score(y_test, y_pred_0)}, \nla matriz de confusion es  {confusion_matrix(y_test, y_pred_0)} \ny vemos el reporte de clasificacion como  {classification_report(y_test, y_pred_0)}')
```

Aunque sea una variable objetivo de clasificacion, podemos utiizar la regresion y tomar el valor que nos entregue como la probabilidad de que sea 1. 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
mod_0_r = RandomForestRegressor(random_state=42)
mod_0_r.fit(X_train, y_train)
y_pred_0_r = mod_0_r.predict(X_test)
print(y_pred_0_r)
print(f'El resultado de nuestro modelo de REGRESION en forma simple es un mean squared error de : {mean_squared_error(y_test, y_pred_0_r)} y un r2 de: {r2_score(y_test, y_pred_0_r)}')
```

De cualquier forma esto solo ha sido un experimento para ver como se comporta la regresion con un data set de este tipo.

Continuamos creando un pipeline para la optimizacion de nuestro modelo.

```python
pipe =Pipeline([
    ('scale', StandardScaler()),
    ('model', RandomForestClassifier(random_state=42))
])
#pipe.get_params()
```

Y un grid search para darle mas profundidad a la optimizacion de hiperparametros

```python

grid = GridSearchCV(
    estimator=pipe,
    param_grid={
    'model__n_estimators': [10, 50, 100],
    'model__max_depth': [None, 5, 10, 20],
    'model__min_samples_split': [2, 3, 5],
    'model__max_features': ['sqrt', 'log2', None]
    },
    cv=7,
    n_jobs=1,
    verbose=2,
    scoring='roc_auc'
    
)
```

```python
grid.fit(X_train, y_train)
```

```python
b_param = grid.best_params_
print(f'Los mejores parametros que pudimos encontrar en esta exploracion han sido:{b_param}')
```

```python
pipe.set_params(**grid.best_params_)
```

```python
pipe.fit(X_train, y_train)

pipe_pred_1 = pipe.predict(X_test)
```

```python
accuracy = accuracy_score(pipe_pred_1, y_test)
print(f"Accuracy en el conjunto de prueba: {accuracy}")
```

Observamos que los valores designados para los hiperparametros no nos acercan a un valor ideal por lo que haremos un grid search random para encontrar el umbral de valores donde nuestro modelo logre un comportamiento optimizado respecto a los valores de default

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
    'n_estimators': [2, 10, 50, 100, 500],
    'max_depth': [None, 2, 5, 10, 20],
    'min_samples_split': [2, 3, 5, 10, 50, 100],
    'max_features': ['sqrt', 'log2', None],
   'criterion': ['entropy', 'gini']

}


```

```python

random_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=param_grid,
    n_iter=1000,  
    cv=7,
    verbose=2,
    n_jobs=1,
    scoring='accuracy'
)
```

```python
random_search.fit(X_train, y_train)
```

```python
best_params = random_search.best_params_
best_score = random_search.best_score_


print("Mejores hiperparámetros:", best_params)
print("Mejor puntuación accuracy:", best_score)
```

Utilizando el random search llegamos a un mejor resultado de con un rango mas amplio de parametros 

Mejores hiperparámetros: {'n_estimators': 100, 'min_samples_split': 50, 'max_features': None, 'max_depth': 5}
Mejor puntuación ROC AUC: 0.8400513439322465

```python
mod= RandomForestClassifier(n_estimators= 50, min_samples_split= 50, max_features= 'sqrt', max_depth= None)
mod.fit(X_train, y_train)
y_pred_rand= mod.predict(X_test)
y_pred_rand
```

```python
accuracy_score(y_test,y_pred_rand)
```

```python
print(classification_report(y_test, y_pred_rand))
```

Utiilzando el random search grid es como logramos mejorar el accuracy del modelo, podemos observar que el calculo de positivos es donde el modelo flaquea. Esta cuestion se debe a un desbalance entre clases, por lo que se revisara la literatura para encontrar mejores enfoques, como agregar peso a las clases o utilizar tecnicas de over sampling o under sampling que nos  permitan balancear las clases y logar eliminar el bias de nuestro modelo a seleccionar valores negativos. 

El modelo opera con una eficiencia relativamente buena, pero unicamente desde un punto de vista educativo.
